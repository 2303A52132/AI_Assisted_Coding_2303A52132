# **Zero-shot:**

 Prompt AI with only the instruction. Write a Python function to determine
whether a given number is prime

```python
import math

def is_prime(n: int) -> bool:
    """Return True if n is a prime number, False otherwise."""
    if n < 2:
        return False
    if n % 2 == 0:
        return n == 2
    r = math.isqrt(n)
    for i in range(3, r + 1, 2):
        if n % i == 0:
            return False
    return True
```

# **One-shot:**

Provide one example: Input: [1, 2, 3, 4], Output: 10 to help AI generate a
function that calculates the sum of elements in a list.

```python
def sum_list(nums: list[int]) -> int:
    """Return the sum of elements in nums."""
    return sum(nums)

# One-shot example
print(sum_list([1, 2, 3, 4]))  # Output: 10
```

# **Few-shot:** 

+ Example 1:
Input: "abc123xyz"
Output: "123"

+ Example 2:
Input: "order56id78"
Output: "5678"

+ Example 3:
Input: "room9floor4"
Output: "94"

to create a function that extracts digits from an
alphanumeric string.

```python
def extract_digits(s: str) -> str:
    """Return concatenated digits from alphanumeric string s."""
    return ''.join(ch for ch in s if ch.isdigit())

# Examples
print(extract_digits("abc123xyz"))    # "123"
print(extract_digits("order56id78"))  # "5678"
print(extract_digits("room9floor4"))  # "94"
```

 # **Compare zero-shot vs few-shot**

prompting for generating a function that counts the
number of vowels in a string.

# **zero-shot**
Generate a function that counts the
number of vowels in a string.

# **Few-shot:**

+ Example 1:
Input: "hello world"
Output: 3

+ Example 2:
Input: "Copilot rocks"
Output: 4

+ Example 3:
Input: "Artificial Intelligence"
Output: 9

Generate a function that counts the
number of vowels in a string.


```python
def count_vowels(s: str) -> int:
    vowels = "aeiouAEIOU"
    count = 0
    for char in s:
        if char in vowels:
            count += 1
    return count

# Example usage:
print(count_vowels("Hello World"))  # Output: 3

def count_vowels(s: str) -> int:
    vowels = set("aeiouAEIOU")  # include both lowercase and uppercase
    return sum(1 for char in s if char in vowels)

# Test cases
print(count_vowels("hello world"))          # Output: 3
print(count_vowels("Copilot rocks"))        # Output: 4
print(count_vowels("Artificial Intelligence"))  # Output: 9
```

## ****Task Description-5****

Use few-shot prompting with 3 sample inputs to generate a function that determines
the minimum of three numbers without using the built-in min() function.

```python
def min_of_three(a: float, b: float, c: float) -> float:
    """Return the minimum of three numbers without using min()."""
    smallest = a
    if b < smallest:
        smallest = b
    if c < smallest:
        smallest = c
    return smallest

# Examples
print(min_of_three(5, 8, 3))    # Output: 3
print(min_of_three(12, 7, 9))   # Output: 7
print(min_of_three(4, 4, 10))   # Output: 4

# Given input
print(min_of_three(15, 20, 11)) # Output: 11
```
